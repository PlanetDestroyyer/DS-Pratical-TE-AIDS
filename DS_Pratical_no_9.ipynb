```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
data = df[['sex','age','survived']]
```

```python
data.head()
```

```python
data = data.dropna(subset=['age'])
```

```python
data.head()
```

```python
sns.boxplot(x='sex' , y='age', hue='survived' , data= data)
plt.title("Distribution of AGE with respect to Sex and Survival")
plt.xlabel('Gender')
plt.ylabel('AGE')
plt.show()
```
