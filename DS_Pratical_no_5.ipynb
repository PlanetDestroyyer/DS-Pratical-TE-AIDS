```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report , recall_score , precision_score , f1_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
url = "https://raw.githubusercontent.com/shivang98/Social-Network-ads-Boost/master/Social_Network_Ads.csv"
```

```python
df = pd.read_csv(url)
```

```python
df['Purchased'].unique()
```

```python
df.drop('User ID',axis=1 ,inplace=True)
```

```python
df.head()
```

```python
encoder = LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
```

```python
df.head()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.35,shuffle=True,random_state=7)
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
```

```python
print("Accuracy of Classifier :")
print(accuracy)
```

```python
cm = confusion_matrix(y_test,y_pred)

sns.heatmap(cm,annot=True)
plt.show()
```

```python
report = classification_report(y_test,y_pred)
print(report)
```

```python
print("Accuracy :")
print(accuracy_score(y_test,y_pred))
```

```python
print("Recall :")
print(recall_score(y_test,y_pred))
```

```python
print("Precision :")
print(precision_score(y_test,y_pred))
```

```python
print("F1 Score :")
print(f1_score(y_test,y_pred))
```

```python
print("Confusion Matrix :")
print(confusion_matrix(y_test,y_pred))
```
