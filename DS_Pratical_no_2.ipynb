```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sample_data = {
    'Name' :['Mohan','Neha','Nitin','Roshan'],
    'Age' : [20,32,53,12],
    'Gender' : ['Male','Female','Male','Male'],
    'Subjects' : ['Math','Science','English','Hindi'],
    'Exam Score': [90, 85, np.nan, 65],
  'Study Hours': [4, 6,np.nan, 5],
  'Attendance Percentage': [80, 95, 85,  60]

}
```

```python
df = pd.DataFrame(sample_data)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```

```python
df['Exam Score'] = df['Exam Score'].fillna(df['Exam Score'].mean())
```

```python
df.head()
```

```python
df['Study Hours'] = df['Study Hours'].fillna(df['Study Hours'].mean())
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.notnull().sum()
```

```python
df.describe()
```

```python
low_bond = 65
high_bond = 95
df['Exam Score'] = np.where(df['Exam Score'] < low_bond, low_bond, df['Exam Score'])

df['Exam Score'] = np.where(df['Exam Score'] > high_bond , high_bond , df['Exam Score'])
```
